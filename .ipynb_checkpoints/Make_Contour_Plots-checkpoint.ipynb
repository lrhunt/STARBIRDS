Import all of the libraries we need for this project

```python
import aplpy as apl
import astropy as ap
import numpy as np
import matplotlib.pyplot as plt
import os
```

Start by defining the NUV, VLA, GBT fits files directories

```python
NUV_DIR='/Users/lucashunt/ASTRODATA/STARBIRDS/NUV_Images/'
VLA_DIR='/Users/lucashunt/ASTRODATA/STARBIRDS/VLA_Mom0/'
GBT_DIR='/Users/lucashunt/ASTRODATA/STARBIRDS/GBT_FITS/'
```

Next we want to get a list of all of the files in each of the directories

```python
GBT_FITS=os.listdir(GBT_DIR)
VLA_FITS=os.listdir(VLA_DIR)
NUV_FITS=os.listdir(NUV_DIR)
#print(GBT_FITS)
#print(VLA_FITS)
gc=apl.FITSFigure(GBT_DIR+'D187_GBT_CUBE_4CH_mom0.fits')
plt.rcParams['axes.facecolor']='black'
gd=apl.FITSFigure(VLA_DIR+'DDO187_NA_X0_P_R.FITS')
gd.recenter(gd.pixel2world(512,512)[0],gd.pixel2world(512,512)[1],radius=0.5)
hdu=ap.io.fits.open(GBT_DIR+'D187_GBT_CUBE_4CH_mom0.fits')
print(hdu[0].header['BMAJ'])
#gc=apl.FITSFigure(VLA_DIR+VLA_FITS[0])
#gd=apl.FITSFigure(GBT_DIR+GBT_FITS[1])
#print(gc.pixel2world(17,17))
#print(VLA_FITS[0],GBT_FITS[1])
#gc.recenter(radius=1)
gc.show_grayscale()
gd.show_grayscale()
gd.show_contour(GBT_DIR+'D187_GBT_CUBE_4CH_mom0.fits',colors='yellow')
gd.add_beam()
gd.add_beam()
#gd.beam[0].show()
print(gd.beam)
gd.beam[1].set_major(float(hdu[0].header['BMAJ']))
gd.beam[1].set_minor(float(hdu[0].header['BMIN']))
gd.beam[1].set_angle(float(hdu[0].header['BPA']))
gd.beam[1].set_color('yellow')
gd.beam[1].set_corner('bottom right')
gd.beam[0].show()
gd.set_title('DDO 187')
gd.save('/Users/lucashunt/ASTRODATA/STARBIRDS/APLPY_Images/Contour_plots/DDO187_CONTOUR.png')
```
