```python
for VLA_FILE in VLA_TO_GBT_DICT:
    print(VLA_FILE)
    gc=apl.FITSFigure(GBT_DIR+VLA_TO_GBT_DICT[VLA_FILE])
    plt.rcParams['axes.facecolor']='black'
    gd=apl.FITSFigure(VLA_DIR+VLA_FILE)
    gd.recenter(gd.pixel2world(512,512)[0],gd.pixel2world(512,512)[1],radius=0.5)
    hdu=ap.io.fits.open(GBT_DIR+VLA_TO_GBT_DICT[VLA_FILE])
    gc.show_grayscale()
    gd.show_grayscale()
    gd.show_contour(GBT_DIR+VLA_TO_GBT_DICT[VLA_FILE],colors='yellow')
    gd.add_beam()
    gd.add_beam()
    gd.beam[1].set_major(float(hdu[0].header['BMAJ']))
    gd.beam[1].set_minor(float(hdu[0].header['BMIN']))
    gd.beam[1].set_angle(float(hdu[0].header['BPA']))
    gd.beam[1].set_color('yellow')
    gd.beam[1].set_corner('bottom right')
    gd.beam[0].show()
    gd.set_title('VLA_FILE')
```
